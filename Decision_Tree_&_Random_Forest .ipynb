```python
import pandas as pd
df=pd.read_csv("titanic-passengers.csv",delimiter=';')
df.head()
```

```python
df.isna().sum()
```

```python
df.Age.where(df.Age>25,0, inplace=True)
```

```python
df
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
df.drop("Cabin",axis=1,inplace=True)
df["Embarked"].fillna("S",inplace=True)
df["Sex"]=df["Sex"].map({'male':1,'female':0})
df["Survived"]=df["Survived"].map({'Yes':1,'No':0})
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df.drop(["PassengerId","Survived","Name","Ticket","Embarked"],axis=1)
y=df["Survived"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
tree1=tree.DecisionTreeClassifier(random_state=0)
tree1.fit(x_train,y_train)
y_pred=tree1.predict(x_test)
print("score:{}".format(accuracy_score(y_test,y_pred)))
```

```python
import graphviz
dot_data=tree.export_graphviz(tree1,out_file=None)
graph=graphviz.Source(dot_data)
graph.render("data")
graph
```

```python
dtree=tree.DecisionTreeClassifier(criterion="gini",splitter="random",max_leaf_nodes=10,min_samples_leaf=5,max_depth=4, random_state=0)
```

```python
dtree.fit(x_train,y_train)
y_pred=dtree.predict(x_test)
print("score:{}".format(accuracy_score(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
clf=RandomForestClassifier(n_estimators=10, random_state=0)
clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_predict))
```

```python
clf=RandomForestClassifier(n_estimators=25,random_state=0)
clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_predict))
```

```python
clf=RandomForestClassifier(n_estimators=25, random_state=0)
clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_predict))
```
